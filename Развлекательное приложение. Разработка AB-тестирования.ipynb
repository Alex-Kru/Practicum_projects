### 2. Подготовка к тесту
При планировании теста необходимо проделать несколько важных шагов:

- Определиться с целевой метрикой.

- Рассчитать необходимый размер выборки.

- Исходя из текущих значений трафика рассчитать необходимую длительность проведения теста.

#### 2.1. Расчёт размера выборки
В рамках курса вы уже рассчитывали размеры выборки и  использовали для этого онлайн-калькулятор. В этом задании предлагаем воспользоваться готовым кодом и рассчитать необходимое для вашего эксперимента количество пользователей.

Для этого установите в коде ниже следующие параметры:

- Уровень значимости — 0.05.

- Вероятность ошибки второго рода — 0.2.

- Мощность теста.

- Минимальный детектируемый эффект, или MDE, — 3%. Обратите внимание, что здесь нужно указать десятичную дробь, а не процент.

При расчёте размера выборки используйте метод `solve_power()` из класса `power.NormalIndPower` модуля `statsmodels.stats`.

Запустите ячейку и изучите полученное значение.

````python
from statsmodels.stats.power import NormalIndPower
from statsmodels.stats.proportion import proportion_effectsize

# Задайте параметры:
alpha = 0.05  # Уровень значимости
beta = 0.2  # Ошибка второго рода, часто 1 - мощность
power = 1-beta  # Мощность теста
p1 = 0.3 # Базовый уровень доли
mde = 0.03*p1  # Минимальный детектируемый эффект
effect_size = proportion_effectsize(p1, p1 + mde)

# Инициализируем класс NormalIndPower
power_analysis = NormalIndPower()

# Расчёт размера выборки
sample_size = power_analysis.solve_power(
    effect_size = effect_size,
    power = power,
    alpha = alpha,
    ratio = 1, # Равномерное распределение выборок
    )

print(f"Необходимый размер выборки для каждой группы: {int(sample_size)}")
````

---    

✅ Здорово, что связана ошибка второго рода и мощность критерия через `power = 1 - beta`. Это важный аспект статистического анализа, который помогает оценить вероятность обнаружения реального эффекта.  

Преимущества такого подхода:

Для расчёта процентной разницы воспользуйтесь формулой:
$$P = 100 \cdot  \frac{|A − B|}{A}$$

````python
# Сохраняем новый датафрейм
sessions_test_part=pd.read_csv('/datasets/sessions_project_test_part.csv')

#проверяем количество уникальных дат в датафрейме
unique_days = sessions_test_part['session_date'].value_counts()
print(f'уникальные дни : {unique_days}')

# Считаем количество уникальных пользователей 
total_users=sessions_test_part['user_id'].nunique()
print(f'Количество уникальных пользователей : {total_users}')

# Считаем количество уникальных пользователей каждой группы за один день
test_group_A=sessions_test_part[sessions_test_part['test_group'] == 'A']['user_id'].nunique() 
test_group_B=sessions_test_part[sessions_test_part['test_group'] == 'B']['user_id'].nunique() 
print(f'Количество пользователей в группе А: {test_group_A}, в группе В: {test_group_B}, сумма: {test_group_A+test_group_B}')

# Считаем процентную разницу в количестве пользователей в группах А и В
percent_A = 100 * test_group_A / total_users
percent_B = 100 * test_group_B / total_users
defferents= 100 * (percent_A-percent_B)/percent_A

print(f"Процентная разница в количестве пользователей между группами: {round(defferents,2)}%")

# Создаём круговую диаграмму, наиболее подходящая для сранения 2-3 показателей
plt.figure(figsize=(4, 4))

sizes = [percent_A, percent_B]
labels = ['Группа A', 'Группа B']

plt.pie(sizes, labels=labels,autopct='%1.1f%%')
plt.title("Разница пользователей групп А и В (%)")
plt.show()
````

#### 4.3. Сравнение доли успешных сессий

Перейдем к анализу ключевой метрики — доле успешных первых сессий.

Используйте созданный на первом шаге задания столбец `good_session` и рассчитайте долю успешных первых сессий для выборок A и B, а также разницу в этом показателе. Полученный вывод отобразите на экране.

````python
#считаем количество всех сессий для каждой группы А и В
n_a=sessions_test[sessions_test['test_group']=='A'].shape[0]
n_b=sessions_test[sessions_test['test_group']=='B'].shape[0]

#считаем количество "хороших" сессий каждой группы А и В
m_a=sessions_test[(sessions_test['test_group']=='A')&(sessions_test['good_session']==1)].shape[0]
m_b=sessions_test[(sessions_test['test_group']=='B')&(sessions_test['good_session']==1)].shape[0]

#расчитываем доли удачных сессий для каждой группы А и В
p_a=m_a/n_a
p_b=m_b/n_b

#расчитываем разницу долей
diff_ratio=(p_b-p_a)*100

print(f'Разница в доле успешных сессий между группами A ({round(p_a*100,1)}%) и B ({round(p_b*100,1)}%) составляет: {round(diff_ratio,1)}%')
````

#### 4.4. Насколько статистически значимо изменение ключевой метрики

На предыдущем шаге вы убедились, что количество успешных сессий в тестовой выборке примерно на 1.1% выше, чем в контрольной, но делать выводы только на основе этого значения будет некорректно. Для принятия решения всегда необходимо отвечать на вопрос: является ли это изменение статистически значимым.

- Используя статистический тест, рассчитайте, является ли изменение в метрике доли успешных сессий статистически значимым.

- Выведите на экран полученное значение p-value и свои выводы о статистической значимости. Напомним, что уровень значимости в эксперименте был выбран на уровне 0.05.

````python
# определить, существует ли значимая разница между долями в двух независимых выборках позволит Z-тест 

from statsmodels.stats.proportion import proportions_ztest

alpha = 0.05 # На каком уровне значимости проверяем гипотезу о равенстве вероятностей

stat_ztest, p_value_ztest = proportions_ztest(
    [m_a, m_b],
    [n_a, n_b],
    alternative='smaller' # так как H_1: p_a < p_b
)

if p_value_ztest > alpha:
    print(f'pvalue={p_value_ztest} > {alpha}')
    print('Нулевая гипотеза не отвергается: разница статистически не значима.')
else:
    print(f'pvalue={p_value_ztest} < {alpha}')
    print('Нулевая гипотеза отвергается: разница статистически значима.')

````
