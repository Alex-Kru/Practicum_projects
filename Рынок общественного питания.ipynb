````python
# изучим пропуски с помощью функции 
def show_missing_stats(tmp0):
    """
    Функция для отображения статистики пропущенных значений в DataFrame.
    """
    missing_stats = pd.DataFrame({
        'Кол-во пропусков': tmp0.isnull().sum(),
        'Доля пропусков': tmp0.isnull().mean()
    })
    missing_stats = missing_stats[missing_stats['Кол-во пропусков'] > 0]
    
    if missing_stats.empty:
        return "Пропусков в данных нет"
    
    # Форматируем при выводе через Styler
    return (missing_stats.style.format({'Доля пропусков': '{:.4f}'}).background_gradient(cmap='coolwarm'))
show_missing_stats(df)
````

````python
# напишем функцию для написания абриавиатуры вместо полного наименования округа
def abrv (district):
    district = district.replace('-', ' ').split()
    abrv = ""
    for word in district:
        abrv = abrv + word[0].upper()
    return abrv

df['district_short'] = df['district'].apply(abrv)
````

````python
# Сделаем сводную таблицу со средними чеками по районам
pivot_district = df_clean.pivot_table(index=['district'],
                                values='middle_avg_bill',
                                aggfunc={'middle_avg_bill':'mean'}).sort_values(by='middle_avg_bill', ascending=False)
print(pivot_district)
````

````python
# Визулизаруем данные на диаграмме с добавлением линии, которая покажет среднее значение по всем районам
# Создадим функцию для анализа данных


def plot_bar_plot(df,groupby,value,aggfunc,title,xlabel,ylabel):
    groupby=df.groupby(groupby).agg({value:aggfunc})
    groupby.plot(kind='bar',
                 title=title,
                 legend=True,
                 xlabel=xlabel,
                 ylabel=ylabel,
                 rot=0,
                 figsize=(13,6))
    
    #расчитаем среднее значение для всех районов
    mean_avg_bill=df.agg({value:aggfunc})[0]
    
    #нанесем линию на график
    plt.axhline(mean_avg_bill, color='red', linestyle='--', linewidth=1,
                label=f'Среднее значение по всем районам {round(mean_avg_bill,2)}')
    
    plt.grid()
    plt.legend()
    plt.show()
````

````python
#применим функцию
title='Распределение среднего чека по районам'
xlabel='районы'
ylabel='средний чек'
plot_bar_plot(df_clean, 'district_short', 'middle_avg_bill', 'mean',title, xlabel, ylabel)
````
