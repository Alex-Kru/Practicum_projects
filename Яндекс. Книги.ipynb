```python
# Определяем выбросы
# Вычисляем квартили Q1 и Q3
Q1_msk = msk['hours'].quantile(0.25)
Q3_msk = msk['hours'].quantile(0.75)

# Вычисляем межквартильный размах IQR
IQR_msk = Q3_msk - Q1_msk

# Определяем верхний порог для выбросов
upper_bound_msk = Q3_msk + 1.5 * IQR_msk
count_upper_bound_msk = (msk['hours'] > upper_bound_msk).sum()

# Определяем выбросы
# Вычисляем квартили Q1 и Q3
Q1_spb = spb['hours'].quantile(0.25)
Q3_spb = spb['hours'].quantile(0.75)

# Вычисляем межквартильный размах IQR
IQR_spb = Q3_spb - Q1_spb

# Определяем верхний порог для выбросов
upper_bound_spb = Q3_spb + 1.5 * IQR_spb
count_upper_bound_spb = (spb['hours'] > upper_bound_spb).sum()


print(f'Количество выбросов от верхней границы для пользователей из Москвы={count_upper_bound_msk}')
print(f'Количество выбросов от верхней границы для пользователей из Санкт-Петербурга={count_upper_bound_spb}')
```

## 2. Проверка гипотезы в Python

Гипотеза звучит так: пользователи из Санкт-Петербурга проводят в среднем больше времени за чтением и прослушиванием книг в приложении, чем пользователи из Москвы. Попробуйте статистически это доказать, используя одностороннюю проверку гипотезы с двумя выборками:

- Нулевая гипотеза H₀: Средняя активность пользователей в часах в двух группах (Москва и Санкт-Петербург) не различается.

- Альтернативная гипотеза H₁: Средняя активность пользователей в Санкт-Петербурге больше, и это различие статистически значимо.

```python
#выделим группы данных
msk_group = msk['hours']
spb_group = spb['hours']

alpha = 0.05
#проведение t-теста
stat_welch_ttest, p_value_welch_ttest = ttest_ind(spb_group, msk_group, equal_var=False, alternative='greater')

print(f'P-value: {p_value_welch_ttest}')


if p_value_welch_ttest > alpha:
    display("Не отвергаем нулевую гипотезу: различие между группами статистически незначимо")
else:
    display("Отвергаем нулевую гипотезу: различие между группами статистически значимо")
```

## 4. Проведите оценку результатов A/B-тестирования:

- Проверьте изменение конверсии подходящим статистическим тестом, учитывая все этапы проверки гипотез.

Сформулируем гипотизы:
H0: Конверсия в группе А и группе В равны
H0: Конверсия в группе А меньше чем в группе В 

```python
merged_total=merged_total.reset_index()
merged_total
```

```python
# определить, существует ли значимая разница между долями в двух независимых выборках позволит Z-тест 

# общее количество пользователей
n_a = merged_total[merged_total['group']=='A']['total_users']
n_b = merged_total[merged_total['group']=='B']['total_users']
print(n_a)
print(n_b)
#количество пользователей совершивших покупку
m_a=merged_total[merged_total['group']=='A']['total_purchase']
m_b=merged_total[merged_total['group']=='B']['total_purchase']
print(m_a)
print(m_b)
```

```python
#расчитываем доли для каждой группы А и В
p_a=m_a/n_a
p_b=m_b/n_b

alpha = 0.05 

stat_ztest, p_value_ztest = proportions_ztest(
    [m_a, m_b],
    [n_a, n_b],
    alternative='smaller' # так как H_1: p_a < p_b
)

if p_value_ztest > alpha:
    print(f'pvalue={p_value_ztest} > {alpha}')
    print('Нулевая гипотеза не отвергается: разница статистически не значима.')
else:
    print(f'pvalue={p_value_ztest} < {alpha}')
    print('Нулевая гипотеза отвергается: разница статистически значима.')
```
